```python
# Build time formatter 

def stringify_int_time(dataframe, column_name):
    
#     dataframe[column_name] = dataframe[column_name].astype(int)
    dataframe[column_name] = dataframe[column_name].apply(str)
    
    # Build 'masks'
    list_of_three = dataframe.loc[dataframe[column_name].apply(lambda x: len(x)==3)]
    list_of_two = dataframe.loc[dataframe[column_name].apply(lambda x: len(x)==2)]
    list_of_one = dataframe.loc[dataframe[column_name].apply(lambda x: len(x)==1)]
    list_of_zero = dataframe.loc[dataframe[column_name].apply(lambda x: len(x)==0)]
    
    # Apply transformation
    dataframe.loc[list_of_three.index, column_name] = '0' + dataframe.loc[list_of_three.index,column_name]
    dataframe.loc[list_of_two.index, column_name] = '00' + dataframe.loc[list_of_two.index,column_name]
    dataframe.loc[list_of_one.index, column_name] = '000' + dataframe.loc[list_of_one.index,column_name]
    dataframe.loc[list_of_zero.index, column_name] = '0000' + dataframe.loc[list_of_zero.index,column_name]
    return dataframe
```

```python
# Build time formatter for float

def stringify_float_time(dataframe, column_name):
    
#     dataframe[column_name] = dataframe[column_name].astype(int)
#     dataframe[column_name] = dataframe[column_name].apply(str)
    
    # Build 'masks'
    list_of_three = dataframe.loc[dataframe[column_name].apply(lambda x: len(x)==3)]
    list_of_two = dataframe.loc[dataframe[column_name].apply(lambda x: len(x)==2)]
    list_of_one = dataframe.loc[dataframe[column_name].apply(lambda x: len(x)==1)]
    list_of_zero = dataframe.loc[dataframe[column_name].apply(lambda x: len(x)==0)]
    
    # Apply transformation
    dataframe.loc[list_of_three.index, column_name] = '0' + dataframe.loc[list_of_three.index,column_name]
    dataframe.loc[list_of_two.index, column_name] = '00' + dataframe.loc[list_of_two.index,column_name]
    dataframe.loc[list_of_one.index, column_name] = '000' + dataframe.loc[list_of_one.index,column_name]
    dataframe.loc[list_of_zero.index, column_name] = '0000' + dataframe.loc[list_of_zero.index,column_name]
    
    # Readjust NaN values
    list_of_nan = dataframe.loc[dataframe[column_name].apply(lambda x: x=='000n')]
    dataframe.loc[list_of_nan.index, column_name] = np.nan
    
    # Readjust 2400 values
    list_of_24 = dataframe.loc[dataframe[column_name].apply(lambda x: x=='2400')]
    dataframe.loc[list_of_24.index, column_name] = '0000'
    return dataframe
```

```python
stringify_float_time(new_flights_df, 'DEPTIME')
stringify_float_time(new_flights_df, 'ARRTIME')
stringify_float_time(new_flights_df, 'WHEELSOFF')
stringify_float_time(new_flights_df, 'WHEELSON')
```

```python
# ##CHECK FLIGHT DATE
def convert_deptime_to_datetime_from_text(row):
    return pd.to_datetime(row['FLIGHTDATE'].strftime("%Y%m%d") + ' ' + row['DEPTIME'][0:2] + ":" +row['DEPTIME'][2:5] )

def convert_arrtime_to_datetime_from_text(row):
    return pd.to_datetime(row['FLIGHTDATE'].strftime("%Y%m%d") + ' ' + row['ARRTIME'][0:2] + ":" +row['ARRTIME'][2:5] )

def convert_wheelsoff_to_datetime_from_text(row):
    return pd.to_datetime(row['FLIGHTDATE'].strftime("%Y%m%d") + ' ' + row['WHEELSOFF'][0:2] + ":" +row['WHEELSOFF'][2:5] )

def convert_wheelson_to_datetime_from_text(row):
    return pd.to_datetime(row['FLIGHTDATE'].strftime("%Y%m%d") + ' ' + row['WHEELSON'][0:2] + ":" +row['WHEELSON'][2:5] )
```

```python
# Convert DEPTIME to datetime
mask = new_flights_df[new_flights_df['DEPTIME'].notnull()]
new_flights_df['DEPTIME'] = new_flights_df.loc[mask.index].apply(convert_deptime_to_datetime_from_text, axis=1) 
```

```python
# Convert ARRTIME to datetime
mask = new_flights_df[new_flights_df['ARRTIME'].notnull()]
new_flights_df['ARRTIME'] = new_flights_df.loc[mask.index].apply(convert_arrtime_to_datetime_from_text, axis=1) 
```

```python
# Convert WHEELSOFF to datetime
mask = new_flights_df[new_flights_df['WHEELSOFF'].notnull()]
new_flights_df['WHEELSOFF'] = new_flights_df.loc[mask.index].apply(convert_wheelsoff_to_datetime_from_text, axis=1) 
```

```python
# Convert WHEELSON to datetime
mask = new_flights_df[new_flights_df['WHEELSON'].notnull()]
new_flights_df['WHEELSON'] = new_flights_df.loc[mask.index].apply(convert_wheelson_to_datetime_from_text, axis=1) 
```

```python
new_flights_df['temp_time_diff'] = new_flights_df['ARRTIME'] - new_flights_df['DEPTIME']
```

```python
# NEXTDAYARR - additional column
```

```python
new_flights_df['NEXTDAYARR'] = np.nan
```

```python
def apply_nextdayarr_formula(row):
    
    next_day_result = np.nan
    
    if pd.notnull(row['temp_time_diff']):
        if row['temp_time_diff']>=pd.Timedelta(0):
            next_day_result = 0
        else:
            if row['temp_time_diff']<pd.Timedelta('P-1DT24H0M0S'):
                next_day_result = 0
                if row['temp_time_diff']<pd.Timedelta('P-1DT20H0M0S'):
                    next_day_result = 1
    return next_day_result

new_flights_df['NEXTDAYARR'] = new_flights_df.apply(apply_nextdayarr_formula, axis=1)
```

```python
new_flights_df.loc[new_flights_df['NEXTDAYARR']==1, 'ARRTIME'] =  new_flights_df['ARRTIME'] + pd.Timedelta(days=1)
new_flights_df.loc[new_flights_df['NEXTDAYARR']==1, 'CRSARRTIME'] =  new_flights_df['CRSARRTIME'] + pd.Timedelta(days=1)
```

```python
new_flights_df.loc[new_flights_df['NEXTDAYARR']==1, 'WHEELSON'] =  new_flights_df['WHEELSON'] + pd.Timedelta(days=1)
```
